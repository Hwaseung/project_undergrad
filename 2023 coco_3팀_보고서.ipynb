## 서론



### 1) 개요

데이콘에서 진행했던 '구직자들을 위한 기업 트렌드 시각화'라는 주제를 구체화하여 '문과생들을 위한 구직 정보 시각화'를 주제로 대시보드를 제작함.

### 2) 기획 의도

대기업 390곳 중 채용 규모를 축소, 중단 하는 곳이 36.7%에 이르는 만큼 취업이 어려워지고 있다. 특히 전부터 취업문이 좁았던 문과생들은 취업이 더욱 힘들어졌다. 따라서 문과생들이 취업을 하는데 도움이 될 수 있도록 구인구직 데이터로 취업 트렌드를 파악하고 공기업과 사이업 데이터를 활용하여 구직 정보를 제공하는 대시보드를 제작하였다.
